```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from Bio import SeqIO
record = SeqIO.read ("/Users/arina.anurieva/Downloads/chr1.fa", "fasta")
seq_real = str(record.seq)[:5000].upper()    # если брать больше нуклеотидов то p-value становится меньше компьютерного нуля
```

```python
nuc_count = {'A': 0, 'T': 0, 'G': 0, 'C': 0}
for i in range(len(seq_real)):
    nuc_count[seq_real[i]] += 1
```

```python
probs = np.array(list(nuc_count.values()))/len(seq_real)
seq_random = ''.join(np.random.choice(['A', 'T', 'G', 'C'], size=len(seq_real), p=probs))
```

```python
O_freq = np.zeros((4,4), dtype=int)
nuc_to_idx = {'A':0, 'C':1, 'G':2, 'T':3}

for k in range(len(seq_real)-1):
    i = nuc_to_idx[seq_real[k]]
    j = nuc_to_idx[seq_real[k+1]]
    O_freq[i,j] += 1
```

```python
print(f"Матрица частот: \n{O_freq}")
```

```python
E_freq = np.zeros((4,4), dtype=int)
idx_to_nuc = {0:'A', 1:'C', 2:'G', 3:'T'}

for i in range(4):
    prob_i = nuc_count[idx_to_nuc[i]]/len(seq_real)
    for j in range(4):
        prob_j = nuc_count[idx_to_nuc[j]]/len(seq_real)
        E_freq[i, j] = prob_i*prob_j*len(seq_real)
```

```python
print(f"Матрица ожидаемых частот: \n{E_freq}")
```

```python
from scipy.stats import chi2

chi_2 = np.sum((O_freq - E_freq)**2 / E_freq)
p_value = chi2.sf(chi_2, 9)

print(f"Значение хи-квадрат {chi_2.round(3)}, p-value = {p_value}")
```

**Вывод:** значение p-value меньше 0.05 — отвергаем гипотезу, нужна марковская цепь первого порядка.
