```python
sites = ["GAGGTAAAC", "TCCGTAAGC", "CAGGTTGGA", "ACAGTCAGC", "TAGGTCAGC", "CAGGTCAGC", "CAGGTCGAT", "CAGGTCAGC", "CAGGTCAGC", "CAGGTTGGC"]
```

### Задание 2.

```python
import Bio as bp
from Bio import motifs
from Bio.Seq import Seq
from Bio import SeqIO
```

```python
sequences = [Seq(site) for site in sites]
motif = motifs.create(sequences)
```

```python
motif.consensus
```

```python
file_path = "/Users/arina.anurieva/Downloads/chr1.fa"
genom = list(SeqIO.parse(file_path, "fasta"))
first_gene = genom[0][:1000000]
```

```python
pssm = motif.pssm
hits_forward = list(pssm.search(first_gene, threshold=5.0, both=False))     # установка both=False отменяет поиск по комплементарной цепочке
                                                                            # мы его делаем отдельно дальше
```

```python
first_gene_rev = first_gene.reverse_complement()
hits_rev = list(pssm.search(first_gene_rev, threshold=5.0, both=False))

# тут нумерация по перевернутой комплементарной цепи, она не совпадает с нумерацией на имеющейся у нас, поэтому меняем:
motif_length = len(motif)
hits_rev_conv = [(len(first_gene) - pos - motif_length, score, '-') 
                      for pos, score in hits_rev]
```

```python
print('Мотивы на прямой цепи ДНК:')
for i in range(len(hits_forward)):
    print(f"Позиция {hits_forward[i][0]}, score {hits_forward[i][1].round(1)}")
```

```python
print('Мотивы на обратной цепи ДНК:')
for i in range(len(hits_rev_conv)-1, -1, -1):
    print(f"Позиция {hits_rev_conv[i][0]}, score {hits_rev_conv[i][1].round(1)}")
```
