```python
import numpy as np
```

```python
sites = ["GAGGTAAAC", "TCCGTAAGC", "CAGGTTGGA", "ACAGTCAGC", "TAGGTCAGC", "CAGGTCAGC", "CAGGTCGAT", "CAGGTCAGC", "CAGGTCAGC", "CAGGTTGGC"]
```

```python
def PFM(sites):
    # Преобразуем список строк в двумерный массив символов
    # Каждая строка разбивается на список символов
    seq_array = np.array([list(seq) for seq in sites])
    
    # Задаём порядок нуклеотидов
    nucleotides = ['A', 'T', 'G', 'C']
    
    # Для каждого нуклеотида считаем, сколько раз он встречается в каждой позиции
    pfm = np.array([(seq_array == nuc).sum(axis=0) for nuc in nucleotides])
    
    return pfm
```

```python
pfm_matrix = PFM(sites)
pfm_matrix
```

```python
def PFM_to_PPM(sites, pfm_matrix, alpha):
    return (pfm_matrix+alpha)/(len(sites) + 4*alpha)
```

```python
ppm_matrix = PFM_to_PPM(sites, pfm_matrix, 0.1)
ppm_matrix
```

```python
def PPM_to_PWM(ppm_matrix, probabilities):
    return np.log2(ppm_matrix / probabilities[:, np.newaxis])
```

```python
pwm_matrix = PPM_to_PWM(ppm_matrix, np.array([0.295, 0.295, 0.205, 0.205]))
pwm_matrix
```

```python
nucleotides = ['A', 'T', 'G', 'C']

max_indices = np.argmax(pwm_matrix, axis=0)
min_indices = np.argmin(pwm_matrix, axis=0)

max_seq = ''.join(nucleotides[i] for i in max_indices)
min_seq = ''.join(nucleotides[i] for i in min_indices)

max_score = np.sum(pwm_matrix[max_indices, range(pwm_matrix.shape[1])])
min_score = np.sum(pwm_matrix[min_indices, range(pwm_matrix.shape[1])])

print(f"Максимальный скор: {max_score:.4f}")
print(f"Последовательность: {max_seq}")
print(f"Минимальный скор: {min_score:.4f}")
print(f"Последовательность: {min_seq}")
```
