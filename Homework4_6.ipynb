```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P = np.array([[0.229077  , 0.2482066 , 0.3830703 , 0.1396461 ],
       [0.25291181, 0.36173045, 0.0921797 , 0.29317804],
       [0.20027913, 0.29134682, 0.34263782, 0.16573622],
       [0.13647521, 0.27736868, 0.39567992, 0.19047619]])
```

```python
Ps = {}
for k in range(1, 1001):
    Ps[k] = np.linalg.matrix_power(P, k)
```

```python
for k in [1, 2, 3, 100, 1000]:
    print(f'Матрица P^{k}: \n{Ps[k]}\n')
```

**Вывод:** для матриц в высокой степени можно видеть, что все строки одинаковы и совпадают со стационарным распределением из прошлой задачи (с точностью до округления).

```python
dist = [np.linalg.norm(Ps[k] - Ps[k-1], 'fro') for k in range(2, 1001)]
k_values = range(2, 1001)

plt.figure(figsize=(10, 6))
plt.plot(k_values, dist, 'b-', linewidth=2, label='∥P$^k$ − P$^{k-1}$∥')
plt.xlabel('k')
plt.ylabel('Норма разности матриц')
plt.title('Сходимость матриц перехода')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(k_values[11:], dist[11:], 'b-', linewidth=2, label='∥P$^k$ − P$^{k-1}$∥')
plt.xlabel('k')
plt.ylabel('Норма разности матриц')
plt.title('Сходимость матриц перехода (кроме первых 10 значений норм)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

**Вывод:** Можно видеть, что после резкого падения в начале, разница норм быстро приходит к очень малым значениям (10^-9) и дальше линейно очень медленно уменьшается.
