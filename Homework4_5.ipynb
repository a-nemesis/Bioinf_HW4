```python
import numpy as np
```

```python
from Bio import SeqIO
record = SeqIO.read ("/Users/arina.anurieva/Downloads/chr1.fa", "fasta")
seq = str(record.seq)[:10000].upper()
```

```python
from collections import Counter

dinuc_counts = Counter()
for i in range(len(seq) - 1):
    dinuc = seq[i:i+2]
    dinuc_counts[dinuc] += 1
```

```python
for dinuc in sorted(dinuc_counts.keys()):
    print(f"{dinuc}: {dinuc_counts[dinuc]}")
```

```python
nucleotides = ['A', 'C', 'G', 'T']
transition = np.zeros((4, 4))

total_from = {base: 0 for base in nucleotides}
for i, start in enumerate(nucleotides):
    for j, end in enumerate(nucleotides):
        dinuc = start + end
        total_from[start] += dinuc_counts.get(dinuc, 0)

for i, start in enumerate(nucleotides):
    for j, end in enumerate(nucleotides):
        dinuc = start + end
        transition[i, j] = dinuc_counts.get(dinuc, 0) / total_from[start]
```

```python
print("Катрица переходов (строка — откуда, столбец — куда)")
print("      A      C      G      T")
for i, base in enumerate(nucleotides):
    row = transition[i]
    print(f"{base}: {row[0]:.4f} {row[1]:.4f} {row[2]:.4f} {row[3]:.4f}")
```

```python
for i, nuc, in enumerate(nucleotides):
    print(f"Вероятность уйти из {nuc} куда-то равна {np.sum(transition[i]).round(1)}")
```

```python
pi = [1, 0, 0, 0]
print(f"Для стартового состояния A — {pi}")
for i in range(7):
    pi = pi@transition
    print(f"—> {pi.round(4)}")
```

```python
pi = [0, 0, 1, 0]
print(f"Для стартового состояния G — {pi}")
for i in range(7):
    pi = pi@transition
    print(f"—> {pi.round(4)}")
```

```python
pi = [0, 0.2, 0.5, 0.3]
print(f"Что-то смешанное — {pi}")
for i in range(7):
    pi = pi@transition
    print(f"—> {pi.round(4)}")
```

```python
nuc_counts = Counter()
for i in range(len(seq) - 1):
    nuc = seq[i:i+1]
    nuc_counts[nuc] += 1
print(f"Распределение нуклеотидов в самом куске:")
for nuc in sorted(nuc_counts.keys()):
    print(f"{nuc}: {nuc_counts[nuc]/len(seq)}")
```

**Вывод:** стационарное состояние марковской цепи совпадает с распределением нуклеотидов в данно участке генома.
